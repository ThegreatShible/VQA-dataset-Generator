## Oriented Object representation of objects in the image
We'll use this representation to easily use the objects in the image


### defining the class

```python
class Object : 
  def __init__(self, name, coordinate, accuracy):
    self.accuracy = accuracy
    self.name  =name
    self.coordinate = coordinate
class Coordinate : 
  def __init__(self, x, y, w, h):
    self.x = x
    self.y = y
    self.h = h
    self.w = w
```

### transforming the String lines into tuples (image_name, object)

```python
def parse_lines(lines):
  def parse_line(line):
    try : 
      l = line.rstrip().split(',')
      image = l[0]
      name = l[1][2:-1]
      accuracy = float(l[2])
      x = float(l[3])
      y = float(l[4])
      w = float(l[5])
      h = float(l[6])
      coordinate = Coordinate(x,y,w,h)
      obj = Object(name, coordinate, accuracy)
      return (image, obj)
    except : 
      return None
  return [parse_line(x) for x in lines if parse_line(x)]
```

## Pretreatement of objects
The pretreatement of objects is done by applying 2 filters. 

*   filter_accuracy : which filters objects by the detection accuracy
*  filter_objects : filter objects by getting only objects listed in all_objects

after filtering, we group by images. We obtain a list of (image_name, list of objects)



```python
#filter based on detection confidence rate 
def filter_accuracy(img_objs):
  return [(img,obj) for (img, obj) in img_objs if obj.accuracy > CONFIDENCE_RATE ]

def filter_objects(img_objs):
  # in the case of NYU Depth V2, all the images are indoor images, so I don't need filtering the image
  return img_objs

#group objects by image. Result is list of (image, objects in that image of type object)
def group_objects(img_objs):
  res = defaultdict(list)
  for v, k in img_objs: res[v].append(k)
  return res

# return lines of the file
def read_lines(file):
  with open(file) as f : 
    lines = f.readlines()
  return [x.strip() for x in lines]
```

## Generate questions 
We are going to generate questions, we use the templates listed above and the objects detected by yolo.

```python

def generate_questions_answers(file, room):
  lines = read_lines(file)
  parsed = parse_lines(lines)
  filtered_objects = filter_accuracy(filter_objects(parsed))
  #unrich all_objects with objects detected by the algorithm, this is done to prevent biases,
  #because when an object is not added in all_objects, we can generate questions with answer "no"
  for im, obj in filtered_objects : 
      if not obj.name in all_objects:
        all_objects.append(obj.name)
  f_imgs, f_objs = zip(*filtered_objects)
  filtered = [x for x in parsed if x[0] in f_imgs]
  questions = []
  grouped = group_objects(filtered)

  for img, objs in grouped.items():
    name_objs = [x.name for x in objs]
    q1 = generate_counting_questions(name_objs)
    q2 = generate_What_room_quetsion(room)
    q3 = generate_detect_questions(name_objs)
    q4 = generate_detect_with_position_questions(objs)
    q5 = generate_count_one_position_questions(objs)
    quests = q1+q2+q3+q4+q5
    questions.append((img, quests))
  return questions


def generate_counting_questions(objects):

  counts = {x:objects.count(x) for x in objects}
  questions_answers = []
  for (obj, count) in counts.items():
    i = random.randint(0,1)
    questions_answers.append((parse(counting_questions[i], [obj]), str(count)))
    questions_answers.append((parse(counting_questions[2], [str(count),obj]),'yes'))
    random_num = generate_number(0,12, [count])
    questions_answers.append((parse(counting_questions[2], [str(random_num), obj]), 'no'))
  objs_idx = get_objects_index(objects)
  fake_questions_answers = []
  for num_fobj in range(2):
    random_number = generate_number(0,len(all_objects)-1, objs_idx)
    fake_obj = all_objects[random_number]
    i = random.randint(0,1)
    fake_questions_answers.append((parse(counting_questions[i], [fake_obj]), 0))
    fake_questions_answers.append((parse(counting_questions[2], ['0',fake_obj]),'yes'))
    random_num = generate_number(1,12, None)
    fake_questions_answers.append((parse(counting_questions[2], [str(random_num), fake_obj]), 'no'))
  return questions_answers + fake_questions_answers
                               
def generate_What_room_quetsion(room):
  rand_number = random.randint(0,1)
  questions_answers = []
  q1= (what_room_questions[rand_number], room)
  q2 = (parse(what_room_questions[2],[room]), 'yes')
  questions_answers.append(q1)
  questions_answers.append(q2)
  #first fake room index in the loop to avoid to pick it again in the second
  first_fake_room_idx = None
  for i in range(2):
    room_index = all_rooms.index(room)
    if not first_fake_room_idx is None:
      fake_room_index = generate_number(0, len(all_rooms)-1, [room_index, first_fake_room_idx])
    else :
      fake_room_index = generate_number(0, len(all_rooms)-1, [room_index])
      first_fake_room_idx = fake_room_index
    fake_room = all_rooms[fake_room_index]
    q = (parse(what_room_questions[2],[fake_room]), 'no')
    questions_answers.append(q)
  return questions_answers
                                  
                                  
def generate_detect_questions(objects):
  rand_number = random.randint(0,1)
  questions = []
  for obj in objects:
    q1 = (parse(detect_objects_questions[rand_number], [obj]), 'yes')
    questions.append(q1)
                                 
  obj_indxs = get_objects_index(objects)
  for i in range(2):
    fake_obj_index = generate_number(0, len(all_objects) -1, obj_indxs)
    fake_obj = all_objects[fake_obj_index]
    q = (parse(detect_objects_questions[rand_number], [fake_obj]), 'no')
    questions.append(q)
  return questions          
 
# objects is the list of objects not their names
def generate_detect_with_position_questions(objects):
  #list of lists
  questions = []
  ordered_objects = left2right(objects)
  if len(ordered_objects) <2 :
    return []
  else :
    o = [x.name for x in objects]
    ordered_objects_name = [ [x.name for x in l] for l in ordered_objects]
    unique_idx = unique_indexes(o)
    unique_index = [get_object_index_in_listOfLists(ordered_objects_name, o[idx]) for idx in unique_idx]
    foreign_objects = list(set(all_objects) - set(o))      
    # Using objects present in the image
    #one element from each list (stage)
    ###selected_list = [random.choice(elem) for elem in ordered_objects]
    for (index, elem) in unique_index:
      
      left_list = ordered_objects_name[0:index]
      if index == len(ordered_objects_name)-1:
        right_list = []
      else:
        right_list = ordered_objects_name[index+1 : len(ordered_objects_name)]
      questions = []
      # if there are elements that are left to this one
      if left_list :
        left_list = [x for y in left_list for x in y]
        left_elem = random.choice(left_list)
        i = random.randint(0,1)
        # i for generating the format of question. The two formats are semantically equivalent
        if i == 0:
          q1 = parse(detect_with_position_questions[i], [elem,'right', left_elem])
          q2 = parse(detect_with_position_questions[i], [left_elem,'left', elem])
          assert elem != left_elem, str(elem) +'  equals  '+ str(left_elem) 
          questions.append((q1, 'true'))
          questions.append((q2, 'true'))
          q3  =  parse(detect_with_position_questions[i], [elem,'left', left_elem])
          q4 =  parse(detect_with_position_questions[i], [left_elem,'right', elem])
          questions.append((q3, 'false'))
          questions.append((q4,'false'))
          random_foreign_object = random.choice(foreign_objects)
          random_position = random.choice(all_positions)
          q6 = parse(detect_with_position_questions[i], [random_foreign_object, random_position, elem ])
          questions.append((q6, 'false'))
        else :
          assert elem != left_elem, str(elem) +'  equals  '+ str(left_elem) 
          q1 = parse(detect_with_position_questions[i], [elem,left_elem,'right'])
          q2 = parse(detect_with_position_questions[i], [left_elem, elem,'left'])
          questions.append((q1, 'true'))
          questions.append((q2, 'true'))
          q3  =  parse(detect_with_position_questions[i], [elem,left_elem,'left'])
          q4 =  parse(detect_with_position_questions[i], [left_elem,elem,'right'])
          questions.append((q3, 'false'))
          questions.append((q4,'false'))
          random_foreign_object = random.choice(foreign_objects)
          random_position = random.choice(all_positions)
          q6 = parse(detect_with_position_questions[i], [random_foreign_object, elem,random_position ])
          questions.append((q6, 'false'))
  return questions                                
    
  
def generate_count_one_position_questions(objects):
  def count_objects(list_of_lists):
    l = [x for elem in list_of_lists for x in elem]
    return sorted({x : l.count(x) for x in set(l)}.items() , key=lambda kv: kv[1] )
  ordered_objects = left2right(objects)
  if len(ordered_objects) < 2 : 
    return []
  else : 
    ordered_objects_name = [ [x.name for x in l] for l in ordered_objects]
    questions = []
    o = [x.name for x in objects]
    #we get only unique objects not to get confusions
    unique_idx = unique_indexes(o)
    #unique index is the index in the list of lists
    unique_index = [get_object_index_in_listOfLists(ordered_objects_name, o[idx]) for idx in unique_idx]
    foreign_objects = list(set(all_objects) - set(o))
    for index, obj_name in unique_index :
      left_list = ordered_objects_name[0:index]
      if index == len(unique_index)-1:
        right_list = []
      else:
        right_list = ordered_objects_name[index+1 : len(ordered_objects_name)]
      positions = []
      if left_list:
        positions.append(('left', left_list))
      if right_list:
        positions.append(('right', right_list))
      for pos, lst in positions:
        dic = count_objects(lst)
        obj, count = random.choice(dic)
        random_object = random.choice(foreign_objects)
        question_chooser = random.randint(0,3)
        if question_chooser <=1 : 
          q1 = parse(count_with_one_position_questions[question_chooser], [obj, pos, obj_name])
          assert obj != obj_name, str(obj) +'  equals  '+ str(obj_name)

          questions.append((q1, str(count)))
          q2 = parse(count_with_one_position_questions[question_chooser], [random_object, pos, obj_name])
          assert obj != obj_name, str(random_object) +'  equals  '+ str(obj_name) 


          questions.append((q2, 0))
        else : 
          q1 = parse(count_with_one_position_questions[question_chooser], [obj, obj_name,pos ])
          assert obj != obj_name, str(obj) +'  equals  '+ str(obj_name) 


          questions.append((q1, str(count)))
          q2 = parse(count_with_one_position_questions[question_chooser], [random_object, obj_name,pos])
          assert obj != obj_name, str(random_object) +'  equals  '+ str(obj_name) 
          questions.append((q2, 0))
  return questions
 
#TODO
#def generate_count_two_positions_questions(objects):
  
#######################################################
def parse(question, args):
  def replace_all(text, dic):
    for i, j in dic.items():
        text = text.replace(i, j)
    return text
  
  def map_optional(optionals):
    dico = {}
    for op in optionals : 
      r = random.randint(0,1)
      if r == 0:
        rep = ''
      else :
        rep = op[1:-1]
      dico[op] = rep
    return OrderedDict(dico)
  replacable_pattern = '\{[a-z]+[0-9]*\}'
  optional_pattern = '\[[a-z]+[0-9]*\]'
  replacables = re.findall(replacable_pattern, question)
  assert(len(replacables) == len(args))
  dic = OrderedDict(zip(replacables, args))
  question = replace_all(question, dic)
  optionals = re.findall(optional_pattern, question)
  question = replace_all(question, map_optional(['[here]']))
  return question
                                  
def left2right(objects):
  
  def is_right_to(a,b):
    if a.x >= b.x + b.w : 
      return True
    if a.x < b.x or (a.x + a.w) < (b.x + b.w):
      return False
    else : 
      intersection = b.x+b.w - a.x
      if (intersection / a.w) <= L2R_FACTOR and (intersection / b.w) <= L2R_FACTOR :
        return True
      else: 
        return False
      
  sorted_objects_by_x = sorted(objects, key = lambda x : x.coordinate.x)
  l = [[sorted_objects_by_x[0]]]
  if len(sorted_objects_by_x) == 1 : 
    return l
  else : 
    #TODO : it's comparing with first elem
    for x in sorted_objects_by_x[1:]:
      done = False      
      i = 0
      while (not done) and (i != len(l)) : 
        if (is_right_to(x.coordinate, l[i][0].coordinate)):
          i+=1
        else : 
          done = True
          l[i].append(x)
      if not done : 
        new_list = [x]
        l.append(new_list)
    return l      
     
def generate_number(bottom, top , not_included):
  i = random.randint(bottom, top)
  if not not_included : 
    return i 
  else : 
    if i in not_included :
      return generate_number(bottom, top, not_included)
    else :
      return i
                                  
# objects is a lise of object (the object not the string)
def get_objects_index(objects):
  return [idx for (idx, obj) in enumerate(objects) if obj in all_objects]
#get indexes of element present only once in the list  
def unique_indexes(objects):
  c = Counter(objects)
  return [i for (i,model) in enumerate(c) if c[model] == 1 ]
#return index in the list of lists of the objects_name and objects_name
def get_object_index_in_listOfLists(list_of_lists, objects_name):
  return [(i, objects_name) for i, lst in enumerate(list_of_lists) if objects_name in lst][0] 
  
```

## generate room name
This function generate room name from directory name, this is a special case applyed to NUY Depth v2 dataset. Don't use it if you are using another dataset

```python
#TODO : the site is down for now i need bathroom
def get_room_name(dir_name):
  if dir_name.lower().startswith('bathroom'):
    return 'bathroom'
  elif dir_name.lower().startswith('basement'):
    return 'basement'
  elif dir_name.lower().startswith('bed'):
    return 'bedroom'
  elif dir_name.lower().startswith('book'):
    return 'bookstore'
  elif dir_name.lower().startswith('cafe'):
    return 'cafe'
  elif dir_name.lower().startswith('class'):
    return 'classroom'
  elif dir_name.lower().startswith('dining'):
    return 'dining room'
  elif dir_name.lower().startswith('furniture'):
    return 'furniture store'
  elif dir_name.lower().startswith('home') :
    return 'home office'
  elif dir_name.lower().startswith('kit'):
    return 'kitchen'
  elif dir_name.lower().startswith('lib'):
    return 'library'
  elif dir_name.lower().startswith('living'):
    return 'living room'
  elif dir_name.lower().startswith('misc'):
    return 'misc'
  elif dir_name.lower().startswith('office kit'):
    return 'office kitchen'
  elif dir_name.lower().startswith('play'):
    return 'playroom'
  elif dir_name.lower().startswith('reception'):
    return 'reception room'
  elif dir_name.lower().startswith('study room'):
    return 'study room'
  #TODO Check this one
  elif dir_name.lower().startswith('stud'):
    return 'study'
  elif dir_name.lower().startswith('office'):
    return 'office'
  raise Exception(dir_name)
```

```python
for file in listdir(descriptions_dir):
  room = get_room_name(file[4:])
  file_path = os.path.join(descriptions_dir, file)
 
  QAs =generate_questions_answers(file_path, room)
  
  output = os.path.join(QAs_result,'QAs.json')
  objs = []
  for k,v in groupby(QAs, lambda x : x[0]) :
    
    json_QAs = [{'question' : q, 'answer' : a} for (q,a) in v]
    obj = {
        'image' : k,
        'QAs' : json_QAs
    }
    objs.append(obj)
    
with open(output, 'w') as f:
  json.dump(objs,f)
    
  
    
  
```
